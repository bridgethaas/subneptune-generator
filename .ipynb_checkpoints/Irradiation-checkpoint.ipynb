```python
##############################################
########     Radius Irradiation   ############
###############################################
import numpy as np
import matplotlib.pyplot as plt
from scipy import loadtxt, optimize
import os
import sys
import os.path
import os.path
import shutil
%matplotlib notebook
from scipy import loadtxt, optimize
from collections import defaultdict
from scipy.interpolate import interp1d

#mpList=[1,5,10]
mpList=[5]
enFracList=[.0025,.005,.0075,.01,.05,.10]
yList = [.24]
zList = [.02]
orbitalList=[.05,.1,.15,.2,.25,.3,.35,.4,.45,.5,.55,.6,.65,.7,.75,.8,.85,.9,.95,1.0]
entropyList = [9.0]
n_frac_list = [.05]  

#mpList = [19.37]
#Iterate again over all the files
#Iterate again over all the files
base_path = '/Users/isaacmalsky/Desktop/LOGS/evolve_'
flux_list = []
rad_list = []

enFrac_list = []
metalicity_list = []
hydrogen_list = []
helium_list = []
envelope_mass = []

for zfrac in zList:
    for enFrac in enFracList:
        for yfrac in yList:
            for entropy in entropyList:
                for orb_sep in orbitalList:
                    for mp in mpList:
                        for n_frac in n_frac_list:              
                            myfile = ''
                            extension = "{}_{}_{}_{}_{}_{}_{}".format(mp,enFrac,yfrac,zfrac,orb_sep,entropy, n_frac)
                            myfile = base_path + extension + ".mod"                     

                            if (os.path.isfile(myfile)):
                                x,y,hydrogen, helium, hydrogen_mass, helium_mass = loadtxt(myfile, unpack=True, skiprows =6, usecols=[0,14,22,24,17,18])
                                y = y / (637.1 * 10 ** 6)
                                earth_flux = (5.67e-5*5700**4 * (.464*6.9598e10 / 1 /1.496e13)**2)*(1-.3)
                                flux_dayside = (5.67e-5*6000**4 * (.464*6.9598e10/orb_sep /1.496e13)**2)*(1-.2) / earth_flux


                                array_type = str(type(x))
                                if (array_type == "<class 'numpy.ndarray'>"):
                                    print (x[-1])
                                    if (x[-1] > 9.698):
                                        #print (hydrogen[-1],helium[-1], (1 - hydrogen[-1] - helium[-1]))

                                        age_radius = interp1d(x,y)
                                        flux_list.append(float(flux_dayside))
                                        rad_list.append(float(age_radius(9.698)))
                                        enFrac_list.append(enFrac)                                  

                                        metalicity_list.append(1 - hydrogen[-1] - helium[-1])
                                        hydrogen_list.append(hydrogen[-1])
                                        helium_list.append(helium[-1])
                                        envelope_mass.append((hydrogen_mass[-1] + helium_mass[-1]) / (hydrogen_mass[0] + helium_mass[0]))
```
