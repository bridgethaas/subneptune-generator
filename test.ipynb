```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


age, star_mass, surface_he, h_mass, he_mass, ne20_mass, env_mass, h_lost, he_lost = np.loadtxt('/Users/imalsky/Desktop/PlanetSuite/LOGS/BAD_mod',
                  unpack=True, skiprows =35, usecols=[0, 2, 5, 6, 7, 8, 50, 51, 52])

from functools import reduce

#print (reduce(lambda x, y: x*y, test))


print ('mass lost h:', (h_mass[0] - h_mass[-1]) * 1.988435e33)
print ('mass lost he:', (he_mass[0] - he_mass[-1]) * 1.988435e33)
print ()

print ('h lost:', sum(h_lost))
print ('he lost', sum(he_lost))
print ('Mesa mass:', (h_mass[0] - h_mass[-1]) * 1.988435e33 + (he_mass[0] - he_mass[-1]) * 1.988435e33)
print ('envelope mass lost:', env_mass[0] - env_mass[-1])
print ('sum of gases lost:', sum(h_lost) + sum(he_lost))

test_lost_h = []
for i in range(len(age) - 1):
    test_lost_h.append(h_mass[i + 1] - h_mass[i])

test_lost_he = []
for i in range(len(age) - 1):
    test_lost_he.append(he_mass[i + 1] - he_mass[i])
    
    
plt.plot(age[:-1], np.asarray(test_lost_h) * -1.988435e33, label='MESA log')
plt.plot(age, h_lost, label='Mass Lost per step')


plt.plot(age[:-1], np.asarray(test_lost_he) * -1.988435e33, label='MESA log he')
plt.plot(age, he_lost, label='Mass Lost per step he')

plt.legend()

```
